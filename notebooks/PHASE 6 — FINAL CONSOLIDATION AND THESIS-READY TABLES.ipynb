```python
# ============================================================
# PHASE 6 — FINAL CONSOLIDATION AND THESIS-READY TABLES
# ============================================================

import warnings
from pathlib import Path

import numpy as np
import pandas as pd

warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", 200)
pd.set_option("display.max_colwidth", 250)

# ============================================================
# PATHS
# ============================================================

import os
from pathlib import Path
import pandas as pd
from google.colab import drive

try:
    drive.mount('/content/drive')
except ValueError:
    print("Drive đã được kết nối từ trước.")

PROJECT_DIR = Path("/content/drive/MyDrive/Project_Protein")

# Existing result folders
PHASE4_RESULT_DIR = PROJECT_DIR / "model" / "phase4_biological_validation_preparation" / "results"
PHASE4_1_RESULT_DIR = PROJECT_DIR / "model" / "phase4_1_biological_validation_enrichment" / "results"
PHASE5_RESULT_DIR = PROJECT_DIR / "model" / "phase5_dnabert2_genomic_foundation" / "results"
PHASE5_4B_RESULT_DIR = PROJECT_DIR / "model" / "phase5_4B_dnabert2_biological_quick_validation" / "results"

# Final Phase 6 output
PHASE6_DIR = PROJECT_DIR / "model" / "phase6_final_consolidation"
CSV_DIR = PHASE6_DIR / "csv"
EXCEL_DIR = PHASE6_DIR / "excel"
MD_DIR = PHASE6_DIR / "markdown"
REPORT_DIR = PHASE6_DIR / "reports"

for folder in [PHASE6_DIR, CSV_DIR, EXCEL_DIR, MD_DIR, REPORT_DIR]:
    folder.mkdir(parents=True, exist_ok=True)

print("Phase 6 output:", PHASE6_DIR)
```

```python
# ============================================================
# HELPER FUNCTIONS
# ============================================================

def read_csv_if_exists(path, required=True):
    path = Path(path)
    if path.exists():
        print("Loaded:", path)
        return pd.read_csv(path)

    msg = f"Missing file: {path}"
    if required:
        raise FileNotFoundError(msg)

    print("Warning:", msg)
    return pd.DataFrame()


def round_numeric_columns(df, digits=4):
    df = df.copy()
    for col in df.columns:
        if pd.api.types.is_float_dtype(df[col]):
            df[col] = df[col].round(digits)
    return df


def save_csv_md(df, name, index=False):
    csv_path = CSV_DIR / f"{name}.csv"
    md_path = MD_DIR / f"{name}.md"

    df.to_csv(csv_path, index=index)

    with open(md_path, "w") as f:
        f.write(df.to_markdown(index=index))

    print("Saved CSV:", csv_path)
    print("Saved MD:", md_path)

    return csv_path, md_path


def clean_text_list(x):
    if pd.isna(x):
        return ""
    return str(x).replace(";", "; ")
```

```python
# ============================================================
# LOAD SOURCE TABLES
# ============================================================

# Phase 4 old master comparison
phase4_master_path = PHASE4_RESULT_DIR / "phase4_0A_final_master_comparison_table.csv"
phase4_master_df = read_csv_if_exists(phase4_master_path)

# Phase 4.1 biological validation
phase4_enrichment_summary_path = PHASE4_1_RESULT_DIR / "phase4_1_enrichment_summary_by_gene_list.csv"
phase4_top50_support_path = PHASE4_1_RESULT_DIR / "phase4_1_top50_curated_support_summary.csv"
phase4_curated_overlap_path = PHASE4_1_RESULT_DIR / "phase4_1_curated_t2d_gene_set_overlap.csv"

phase4_enrichment_summary_df = read_csv_if_exists(phase4_enrichment_summary_path)
phase4_top50_support_df = read_csv_if_exists(phase4_top50_support_path)
phase4_curated_overlap_df = read_csv_if_exists(phase4_curated_overlap_path)

# Phase 5 master comparison
phase5_master_path = PHASE5_RESULT_DIR / "phase5_4_master_comparison_with_dnabert2.csv"
phase5_master_df = read_csv_if_exists(phase5_master_path)

# Phase 5.4B quick validation
phase5_vs_phase3_metric_path = PHASE5_4B_RESULT_DIR / "phase5_4B_phase5_vs_phase3_metric_delta.csv"
phase5_bio_comparison_path = PHASE5_4B_RESULT_DIR / "phase5_4B_phase5_vs_phase3_biological_validation_comparison.csv"
phase5_top50_overlap_path = PHASE5_4B_RESULT_DIR / "phase5_4B_top50_overlap_phase5_vs_phase3.csv"
phase5_error_group_path = PHASE5_4B_RESULT_DIR / "phase5_4B_phase5_vs_phase3_error_group_summary.csv"
phase5_enrichment_summary_path = PHASE5_4B_RESULT_DIR / "phase5_4B_enrichment_summary_by_gene_list.csv"
phase5_top50_support_path = PHASE5_4B_RESULT_DIR / "phase5_4B_phase5_top50_curated_support_summary.csv"
phase5_curated_overlap_path = PHASE5_4B_RESULT_DIR / "phase5_4B_curated_t2d_gene_set_overlap.csv"

phase5_vs_phase3_metric_df = read_csv_if_exists(phase5_vs_phase3_metric_path)
phase5_bio_comparison_df = read_csv_if_exists(phase5_bio_comparison_path)
phase5_top50_overlap_df = read_csv_if_exists(phase5_top50_overlap_path)
phase5_error_group_df = read_csv_if_exists(phase5_error_group_path)
phase5_enrichment_summary_df = read_csv_if_exists(phase5_enrichment_summary_path)
phase5_top50_support_df = read_csv_if_exists(phase5_top50_support_path)
phase5_curated_overlap_df = read_csv_if_exists(phase5_curated_overlap_path)

print("All source tables loaded.")
```

```python
# ============================================================
# TABLE 1 — FINAL MASTER COMPARISON TABLE V2
# ============================================================

# Use Phase 5 master as base because it already includes Phase 1-5 key models.
final_master_v2_df = phase5_master_df.copy()

# Standardize columns if needed
expected_cols = [
    "phase",
    "branch",
    "representation",
    "embedding_policy",
    "feature_type",
    "model",
    "test_roc_auc",
    "test_pr_auc",
    "test_f1",
    "test_mcc",
    "note"
]

final_master_v2_df = final_master_v2_df[
    [col for col in expected_cols if col in final_master_v2_df.columns]
].copy()

final_master_v2_df = final_master_v2_df.sort_values(
    by=["test_roc_auc", "test_pr_auc", "test_mcc"],
    ascending=False
).reset_index(drop=True)

final_master_v2_df.insert(
    0,
    "rank_by_roc_auc",
    np.arange(1, len(final_master_v2_df) + 1)
)

# Add role column
def assign_model_role(row):
    phase = str(row.get("phase", ""))
    representation = str(row.get("representation", ""))

    if phase == "3.1" and "K3" in representation:
        return "official_biologically_validated_model"

    if phase == "5.3" and "DNABERT" in representation:
        return "best_roc_auc_extension_model"

    if phase == "3.1" and "Protein" in str(row.get("branch", "")):
        return "best_protein_only_shared_split_model"

    if phase == "5.2":
        return "genomic_foundation_extension"

    if phase == "2.1":
        return "handcrafted_genomic_baseline"

    if phase == "1.2E":
        return "protein_foundation_baseline"

    return "supporting_baseline"

final_master_v2_df["final_role"] = final_master_v2_df.apply(assign_model_role, axis=1)

final_master_v2_df = round_numeric_columns(final_master_v2_df, 4)

display(final_master_v2_df)

save_csv_md(
    final_master_v2_df,
    "table1_final_master_comparison_v2"
)
```

```python
# ============================================================
# TABLE 2 — FINAL DECISION TABLE
# ============================================================

# From Phase 5.4B metric delta
decision_metrics_df = phase5_vs_phase3_metric_df.copy()

# From Phase 5.4B biological comparison
bio_df = phase5_bio_comparison_df.copy()

# Normalize model labels to merge manually
phase3_metric_row = decision_metrics_df[
    decision_metrics_df["model_name"].str.contains("Phase 3.1", case=False, na=False)
].iloc[0]

phase5_metric_row = decision_metrics_df[
    decision_metrics_df["model_name"].str.contains("Phase 5.3", case=False, na=False)
].iloc[0]

phase3_bio_row = bio_df[
    bio_df["model"].str.contains("Phase 3.1", case=False, na=False)
].iloc[0]

phase5_bio_row = bio_df[
    bio_df["model"].str.contains("Phase 5.3", case=False, na=False)
].iloc[0]

decision_records = [
    {
        "model_role": "Official biologically validated model",
        "phase": "3.1",
        "model": "Combined ProtBERT-SW + K3/K4/Basic SVM RBF",
        "genomic_representation": "Handcrafted K3/K4/Basic TSS-proximal regulatory features",
        "test_roc_auc": phase3_metric_row["roc_auc"],
        "test_pr_auc": phase3_metric_row["pr_auc"],
        "test_f1": phase3_metric_row["f1"],
        "test_mcc": phase3_metric_row["mcc"],
        "top50_curated_support": f"{int(phase3_bio_row['top50_curated_supported_n'])}/50 ({phase3_bio_row['top50_curated_supported_pct']:.1f}%)",
        "top50_significant_terms": int(phase3_bio_row["top50_significant_terms"]),
        "top50_t2d_relevant_terms": int(phase3_bio_row["top50_t2d_relevant_terms"]),
        "top_t2d_relevant_term": phase3_bio_row["top_t2d_relevant_term"],
        "decision": "Retain as main biologically validated model"
    },
    {
        "model_role": "Best ROC-AUC extension model",
        "phase": "5.3",
        "model": "Combined ProtBERT-SW + DNABERT-2 SVM RBF",
        "genomic_representation": "DNABERT-2 genomic foundation embedding",
        "test_roc_auc": phase5_metric_row["roc_auc"],
        "test_pr_auc": phase5_metric_row["pr_auc"],
        "test_f1": phase5_metric_row["f1"],
        "test_mcc": phase5_metric_row["mcc"],
        "top50_curated_support": f"{int(phase5_bio_row['top50_curated_supported_n'])}/50 ({phase5_bio_row['top50_curated_supported_pct']:.1f}%)",
        "top50_significant_terms": int(phase5_bio_row["top50_significant_terms"]),
        "top50_t2d_relevant_terms": int(phase5_bio_row["top50_t2d_relevant_terms"]),
        "top_t2d_relevant_term": phase5_bio_row["top_t2d_relevant_term"],
        "decision": "Report as DNABERT-2 extension with best ROC-AUC"
    }
]

final_decision_df = pd.DataFrame(decision_records)
final_decision_df = round_numeric_columns(final_decision_df, 4)

display(final_decision_df)

save_csv_md(
    final_decision_df,
    "table2_final_model_decision_phase3_vs_phase5"
)
```
