```python
import pandas as pd
import re

def split_mapped_genes(value):
    if pd.isna(value):
        return []
    
    value = str(value).strip()
    if not value:
        return []
    
    parts = re.split(r"\s*,\s*|\s+-\s+", value)
    return [p.strip() for p in parts if p.strip()]

gwas_mapped_expanded = gwas_strict_t2d.copy()

gwas_mapped_expanded["mapped_gene_clean"] = (
    gwas_mapped_expanded["MAPPED_GENE"]
    .apply(split_mapped_genes)
)

gwas_mapped_expanded = (
    gwas_mapped_expanded
    .explode("mapped_gene_clean")
    .dropna(subset=["mapped_gene_clean"])
)

gwas_mapped_expanded["mapped_gene_clean"] = (
    gwas_mapped_expanded["mapped_gene_clean"]
    .astype(str)
    .str.strip()
)

gwas_mapped_expanded = gwas_mapped_expanded[
    gwas_mapped_expanded["mapped_gene_clean"] != ""
].copy()

gwas_mapped_expanded.shape
```

```python
gwas_mapped_expanded.head()
```

```python
def split_reported_genes(value):
    if pd.isna(value):
        return []
    
    value = str(value).strip()
    if not value:
        return []
    
    parts = re.split(r"\s*,\s*", value)
    
    bad_values = {"", "NR", "intergenic"}
    return [p.strip() for p in parts if p.strip() not in bad_values]

gwas_reported_expanded = gwas_strict_t2d.copy()

gwas_reported_expanded["reported_gene_clean"] = (
    gwas_reported_expanded["REPORTED GENE(S)"]
    .apply(split_reported_genes)
)

gwas_reported_expanded = (
    gwas_reported_expanded
    .explode("reported_gene_clean")
    .dropna(subset=["reported_gene_clean"])
)

reported_gene_counts_for_join = (
    gwas_reported_expanded
    .groupby("reported_gene_clean")
    .size()
    .reset_index(name="gwas_reported_gene_mentions")
)

reported_gene_counts_for_join.head()
```

```python
gwas_gene_agg = (
    gwas_mapped_expanded
    .groupby("mapped_gene_clean")
    .agg(
        gwas_n_association_mentions=("mapped_gene_clean", "size"),
        gwas_n_unique_snps=("SNPS", "nunique"),
        gwas_n_unique_studies=("STUDY ACCESSION", "nunique"),
        gwas_n_unique_pubmed_ids=("PUBMEDID", "nunique"),
        gwas_n_unique_regions=("REGION", "nunique"),
        gwas_max_pvalue_mlog=("PVALUE_MLOG", "max"),
    )
    .reset_index()
)

gwas_gene_agg = gwas_gene_agg.merge(
    reported_gene_counts_for_join,
    left_on="mapped_gene_clean",
    right_on="reported_gene_clean",
    how="left"
)

gwas_gene_agg["gwas_has_reported_gene_support"] = (
    gwas_gene_agg["gwas_reported_gene_mentions"]
    .notna()
    .astype(int)
)

gwas_gene_agg["gwas_reported_gene_mentions"] = (
    gwas_gene_agg["gwas_reported_gene_mentions"]
    .fillna(0)
    .astype(int)
)

gwas_gene_agg = gwas_gene_agg.drop(
    columns=["reported_gene_clean"]
)

gwas_gene_agg.head()
```

```python
PRIMARY_POS_THRESHOLD = 0.30
HIGH_CONF_THRESHOLD = 0.50

positive_master = df_clean[
    df_clean["association_score"] >= PRIMARY_POS_THRESHOLD
].copy()

positive_master["label"] = 1
positive_master["label_name"] = "T2D_positive"
positive_master["label_source"] = "Open Targets Platform"
positive_master["label_rule"] = "OpenTargets_association_score_ge_0.30"
positive_master["label_version"] = "Positive_Label_v1"
positive_master["disease_name"] = "type 2 diabetes mellitus"
positive_master["disease_id"] = "MONDO_0005148"

positive_master["positive_tier"] = positive_master[
    "association_score"
].apply(
    lambda x: "Tier_A_high_confidence_ge_0.50"
    if x >= HIGH_CONF_THRESHOLD
    else "Tier_B_strong_positive_0.30_to_0.50"
)
```

```python
ot_evidence_map = {
    "ds_gwas_credible_sets": "has_ot_gwas_credible_sets",
    "ds_clinical_precedence": "has_ot_clinical_precedence",
    "ds_europepmc": "has_ot_europepmc",
    "ds_gene_burden": "has_ot_gene_burden",
    "ds_eva": "has_ot_eva",
}

for score_col, flag_col in ot_evidence_map.items():
    positive_master[flag_col] = (
        positive_master[score_col].notna().astype(int)
    )

all_ds_cols = [c for c in positive_master.columns if c.startswith("ds_")]

positive_master["n_ot_evidence_datasources_present"] = (
    positive_master[all_ds_cols]
    .notna()
    .sum(axis=1)
)
```

```python
positive_master = positive_master.merge(
    gwas_gene_agg,
    left_on="target_symbol",
    right_on="mapped_gene_clean",
    how="left"
)

positive_master["has_gwas_catalog_strict_t2d_mapping"] = (
    positive_master["mapped_gene_clean"]
    .notna()
    .astype(int)
)

positive_master = positive_master.drop(
    columns=["mapped_gene_clean"]
)
```

```python
gwas_count_cols = [
    "gwas_n_association_mentions",
    "gwas_n_unique_snps",
    "gwas_n_unique_studies",
    "gwas_n_unique_pubmed_ids",
    "gwas_n_unique_regions",
    "gwas_reported_gene_mentions",
]

for col in gwas_count_cols:
    positive_master[col] = positive_master[col].fillna(0).astype(int)

positive_master["gwas_has_reported_gene_support"] = (
    positive_master["gwas_has_reported_gene_support"]
    .fillna(0)
    .astype(int)
)
```
